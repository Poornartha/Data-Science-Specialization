```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split as tts
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
```

```python
X_train, X_test, y_train, y_test = tts(data.data, data.target, stratify=data.target, random_state=66)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
test_accuracy = []
train_accuracy = []
neighbour_settings = range(1, 11)
from sklearn.neighbors import KNeighborsClassifier
for n_neighbors in neighbour_settings:
    # Model:
    clf = KNeighborsClassifier(n_neighbors=n_neighbors)
    clf.fit(X_train, y_train)
    # Record Training Set Accuracy: 
    train_accuracy.append(clf.score(X_train, y_train))
    # Test Accuracy:
    test_accuracy.append(clf.score(X_test, y_test))

print(test_accuracy)
print(train_accuracy)
```

```python
# Plotting:
plt.plot(neighbour_settings, train_accuracy, label="training accuracy")
plt.plot(neighbour_settings, test_accuracy, label="test accuracy")
plt.ylabel("Accuracy")
plt.xlabel("n_neighbors")
plt.legend()
```
