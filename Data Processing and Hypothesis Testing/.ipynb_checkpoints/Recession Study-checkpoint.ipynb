```python
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', -1)
```

```python
data = pd.read_table('university_towns.txt', sep="\n")
data_arr = data.values
data_arr
```

```python
state = ['Alabama[edit]']
places = []
t = 0
for i in data_arr:
    if '[edit]' in str(i):
        state.append(str(i[0]))
        t += 1
    else:
        places.append([str(i[0]), state[t]])
places
```

```python
places = pd.Series(places)
```

```python
df = pd.DataFrame(places, columns =['State', 'Region Name'])
df
```
